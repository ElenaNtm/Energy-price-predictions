#Loop the automated URL and import the excel files to the dataframe

```python
base_url = "https://www.enexgroup.gr/documents/20126/200034/"
```

```python
start_date = datetime(2022, 12, 29)
end_date = datetime(2023, 7, 2) #past from that we have no information on the page
date_format = "%Y%m%d"
```

```python
current_date = start_date
```

```python
# Initialize an empty DataFrame
df = pd.DataFrame()
```

```python
while current_date <= end_date:
    date_str = current_date.strftime(date_format)
    url = f"{base_url}{date_str}_EL-DAM_AggrCurves_EN_v01.xlsx"

    try:
        # Read the XLSX file into a pandas DataFrame
        df1 = pd.read_excel(url)
        df = pd.concat([df, df1])
    except HTTPError as e:
        print(f"Error accessing URL: {url}")
        print(e)

    # Increment to the next date
    current_date += timedelta(days=1)
```
