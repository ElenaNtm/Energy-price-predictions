```python
#Base url
xlsx_path = 'https://www.enexgroup.gr/documents/20126/314344/20220103_DER_DOL_EN_v01.xlsx'
df = pd.read_excel(xlsx_path)#Specify the sheet, so that it does not import all the information from the excel file
df.head()
```

```python
base_url = "https://www.enexgroup.gr/documents/20126/314344/"
```

```python
start_date = datetime(2022, 1, 4)
end_date = datetime(2022, 8, 31) #past from that we have no information on the page
date_format = "%Y%m%d"
```

```python
current_date = start_date
while current_date <= end_date:
    if current_date.weekday() not in [5, 6]:
        date_str = current_date.strftime(date_format)
        url = f"{base_url}{date_str}_DER_DOL_EN_v01.xlsx"

        try:
            # Try to download the file from the current URL
            response = requests.get(url)

            # Assuming you save the downloaded file locally before reading it
            xlsx_path = f"{date_str}_DER_DOL_EN_v01.xlsx"
            with open(xlsx_path, 'wb') as f:
                f.write(response.content)

            # Read the XLSX file into a pandas DataFrame, specifying the engine (openpyxl or xlrd)
            try:
                df1 = pd.read_excel(xlsx_path, engine='openpyxl')
            except zipfile.BadZipFile as e:
                print(f"Error: {e}. The file for {date_str} is not a valid Excel file.")
                # Optionally, you can delete the invalid file using os.remove(xlsx_path) here.
                # Otherwise, it will be kept on the local storage.

            df = pd.concat([df, df1])

        except requests.exceptions.RequestException as e:
            print(f"Error while downloading file for {date_str}: {e}")
        except FileNotFoundError as e:
            print(f"Error while reading file for {date_str}: {e}")

    # Increment to the next date
    current_date += timedelta(days=1)
```
